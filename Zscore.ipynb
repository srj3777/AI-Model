```python
import numpy as np

data = np.array([10, 12, 15, 18, 20,37,29,32])
```

```python
mean = np.mean(data)
print(mean)
std = np.std(data)
print(std)

#use mathmetical formul to caluclate Zscores
z_scores = (data - mean) / std
print(z_scores)
```

```python
##Zscore using using SciPy
from scipy.stats import zscore

z_scores_inbuilt = zscore(data)

print(z_scores_inbuilt)
```

```python
##to calculte zscore on titanic train data

import pandas as pd
from scipy.stats import zscore

# Load data
df = pd.read_csv("/content/drive/MyDrive/LAB/train.csv")
df
# Applicable numeric columns
num_cols = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Calculate z-scores
zscore_df = df[num_cols].apply(zscore, nan_policy='omit') ##omitting null values



## to copy zscore to original df
df[num_cols] = zscore_df

print(df.head())


```

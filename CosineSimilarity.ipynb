```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("/content/sample_data/QuestionAnswer.csv")
```

```python
# 1. Read CSV file
questions = data["question"].astype(str).tolist()
answers = data["answer"].astype(str).tolist()
print(questions)
print(answers)
```

```python
# 2. Take user input

user_question = input("Enter your question: ")
```

```python
# 3. Convert to TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(questions + [user_question])
print(tfidf_matrix)
```

```python
 #4. Compute cosine similarity
print("user Vector is:")
print(tfidf_matrix[-1].toarray())
print("Question Vector is:")
print(tfidf_matrix[:-1].toarray())

similarities_score = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])
print(similarities_score)
```

```python
# 5. Get best match
best_match_index = similarities_score.argmax()
print(best_match_index)
best_score = similarities_score[0, best_match_index]
print(best_score)
```

```python
#Display BestScore
if best_score > 0.4:
    print("Best Match: ", answers[best_match_index])
```
